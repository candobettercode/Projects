### Histograms

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa['Overall'], bins = bins,color='#e831dc')
plt.xticks(np.arange(0,110,10).tolist())
plt.yticks(np.arange(0,11000,1000).tolist())
plt.ylabel('No. of Players')
plt.xlabel('Skill Levels')
plt.title('Distributon of Players skill in FIFA 2018')

plt.show()
```

```python
fifa['Weight']
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
```

```python
fifa.Weight
```

```python
fifa.Weight[0]
```

```python
light = fifa.loc[fifa.Weight <150].count()[0]
medium = fifa.loc[(fifa.Weight >=150) & (fifa.Weight <175)].count()[0]
heavy = fifa[fifa.Weight >=175].count()[0]
```

```python
heavy
```

```python
medium
```

```python
light
```

```python
weight = [heavy,medium,light]
lab = ['Over 175','150-175','Under 150']
```

```python
weight
```

```python
explode = (.1,0,0)
plt.title('Weigh Distribution of FIFA Playes (in lbs)')
plt.style.use('ggplot')
plt.pie(weight,labels = lab, autopct='%.2f%%',pctdistance=0.7,explode=explode)
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
```

```python
barcelona
```

```python
madrid = fifa.loc[fifa.Club =='Real Madrid']['Overall']
```

```python
madrid
```

```python
plt.style.use('default')

plt.figure(figsize=(6,8))

lab = ['FC Barcelona','Read Madrid']
plt.boxplot([barcelona,madrid], labels = lab)
plt.title('Professional Soccer Team Comparision')
plt.ylabel('FIFA overall Rating')
plt.show()
```

```python
boxes = plt.boxplot([barcelona,madrid], labels = lab, patch_artist=True,medianprops={'linewidth':3})
for box in boxes['boxes']: 
    #set fill color
    box.set(color='pink')
    #Set edge color
    box.set(color ='red',linewidth=3)
    
```

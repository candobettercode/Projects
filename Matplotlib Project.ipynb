```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
plt.plot(gas.Year,gas.USA)
plt.plot(gas['Year'],gas.Canada)

plt.show()
```

```python
plt.title('Gas Prices over time (in USD)',fontdict={'fontweight':'bold','fontsize':16})

#plt.figure(figsize=(3,2))

plt.plot(gas['Year'],gas['Canada'],'b.-', label = 'Canada')
plt.plot(gas['Year'],gas['Japan'],'r.-', label = 'Japan')
plt.plot(gas['Year'],gas['France'],'g.-',label = 'France')
plt.plot(gas['Year'],gas['Australia'],'y.-',label = 'Australia')

plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()
plt.show()
```

```python
for country in gas:
    if country != 'Year':
        #print (country)
        plt.plot(gas.Year,gas[country],marker='.')
plt.legend()
plt.xticks(gas.Year[::3])
plt.show()
```

```python
country_to_look_at = ['UK','Mexico','Japan','Italy']
for country in country_to_look_at:
    if country != 'Year':
        #print (country)
        plt.plot(gas.Year,gas[country],marker='.',label = country)
plt.legend()
plt.xticks(gas.Year[::3])
plt.show()
```
